```python
import math
import matplotlib.pyplot as plt
import numpy as np

# 定义函数 f(x) = cos(x) - x
def f(x):
    return math.cos(x) - x

# 二分法实现
def bisection_method(f, a, b, tol=1e-5, max_iter=100):
    """
    使用二分法求解方程 f(x) = 0 的根。

    参数：
    f      : 要求解的函数
    a, b   : 初始区间 [a, b]
    tol    : 精度要求，默认值为 1e-5
    max_iter: 最大迭代次数，默认值为 100

    返回：
    root   : 近似解
    iter_count: 实际迭代次数
    """
    if f(a) * f(b) >= 0:
        print("初始区间 [a, b] 不满足 f(a) * f(b) < 0 条件！")
        return None, 0

    iter_count = 0
    while (b - a) / 2.0 > tol:
        c = (a + b) / 2.0  # 计算中点
        iter_count += 1
        if f(c) == 0:  # 如果找到根，返回结果
            break
        elif f(a) * f(c) < 0:  # 根在左半区间
            b = c
        else:  # 根在右半区间
            a = c

        # 超过最大迭代次数
        if iter_count >= max_iter:
            print("达到最大迭代次数！")
            break

    return (a + b) / 2.0, iter_count

# 测试二分法
a = 0  # 区间左端点
b = 1  # 区间右端点
root, iterations = bisection_method(f, a, b)

print(f"求得的根为: {root}")
print(f"迭代次数: {iterations}")

# 可视化结果
x_vals = np.linspace(0, 1, 400)
y_vals = np.cos(x_vals) - x_vals

plt.plot(x_vals, y_vals, label=r'$f(x) = \cos(x) - x$')
plt.axhline(0, color='black',linewidth=1)
plt.axvline(root, color='red', linestyle='--', label=f'Root: {root:.5f}')
plt.title('Graph of f(x) = cos(x) - x')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
